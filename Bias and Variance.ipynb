```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression, Lasso
import datetime
from statistics import mean
```

```python
#importing datasets
df= pd.read_csv("weatherHistory.csv")
```

```python
df
```

```python
df['Date']= pd.to_datetime(df['Formatted Date'], utc= True)
df['Month']=df['Date'].dt.month
df
```

```python
df=df.drop(['Formatted Date','Daily Summary','Pressure (millibars)','Wind Speed (km/h)','Humidity','Loud Cover','Date'], axis=1)
df
```

```python
label_encoder=preprocessing.LabelEncoder()
df['Summary']= label_encoder.fit_transform(df['Summary'])
df['Summary'].unique()
df
```

```python
df['Precip Type']=df['Precip Type'].replace(['rain','snow'],[0,1])
df
```

```python
df.isna().sum()
```

```python
df['Precip Type']= df['Precip Type'].replace(np.nan,0)
df.isna().sum()
```

```python
y=df ['Apparent Temperature (C)']
df=df.drop('Apparent Temperature (C)', axis=1)
print(df)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test =train_test_split(df,y,test_size=0.2, random_state=0)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
Variance=np.var(pred)
Variance
```

```python
SSE=np.mean((np.mean(pred)-y)**2)
Bias=SSE-Variance
Bias
```
